# **------------------Page Segmentation Mode ------------------**



* Page Segmnentation mode represents if the book page images are in block of text or single line text or any other format of text.
* This modes helps the model to understand whether to conside the image text as a complete block or not



```python
!tesseract --help-psm
```

```python

kan_text_book = cv2.imread("/content/kann-book text.jpeg")
rgb_kan_text_book = cv2.cvtColor(kan_text_book, cv2.COLOR_BGR2RGB)
cv2_imshow(rgb_kan_text_book)
```

```python
config_tesseract = '--tessdata-dir tessdata --psm 6'

text_1 = pytesseract.image_to_string(rgb_kan_text_book, lang='kan', config='--psm 6')

print(text_1)
```

```python
eng_text_book = cv2.imread("/content/eng-book img.jpg")
rgb_eng_text_book = cv2.cvtColor(eng_text_book, cv2.COLOR_BGR2RGB)
cv2_imshow(rgb_eng_text_book)
```
